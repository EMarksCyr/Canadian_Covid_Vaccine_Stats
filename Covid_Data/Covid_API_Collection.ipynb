```python
report_url = "https://api.covid19tracker.ca/reports"
```

```python
report = requests.get(report_url)
report.text
```

```python
total_report = report.json()
total_report
```

```python
data = total_report["data"]
data
```

```python
data[0]["date"]
```

```python
covid_data = []
```

```python
for daily_report in data:
    date = daily_report["date"]
    change_cases = daily_report["change_cases"]
    change_fatalities = daily_report["change_fatalities"]
    change_tests = daily_report["change_tests"]
    change_hospitalizations = daily_report["change_hospitalizations"]
    change_criticals = daily_report["change_criticals"]
    change_recoveries = daily_report["change_recoveries"]
    change_vaccinations = daily_report["change_vaccinations"]
    change_vaccinated = daily_report["change_vaccinated"]
    change_vaccines_distributed = daily_report["change_vaccines_distributed"]
    total_cases = daily_report["total_cases"]
    total_fatalities = daily_report["total_fatalities"]
    total_tests = daily_report["total_tests"]
    total_hospitalizations = daily_report["total_hospitalizations"]
    total_criticals = daily_report["total_criticals"]
    total_recoveries = daily_report["total_recoveries"]
    total_vaccinations = daily_report["total_vaccinations"]
    total_vaccinated = daily_report["total_vaccinated"]
    total_vaccines_distributed  = daily_report["total_vaccines_distributed"]
    covid_data.append({"Date": date,
                       "Change in Cases": change_cases,
                       "Change in Fatalities": change_fatalities,
                       "Change in Tests": change_tests,
                       "Change in Hospitalizations": change_hospitalizations,
                       "Change in Criticals": change_criticals,
                       "Change in Recoveries": change_recoveries,
                       "Change in Vaccinations": change_vaccinations,
                       "Change in Vaccinated": change_vaccinated,
                       "Change in Vaccines Distributed": change_vaccines_distributed,
                       "Total Cases": total_cases,
                       "Total Fatalities": total_fatalities,
                       "Total Tests": total_tests,
                       "Total Hospitalizations": total_hospitalizations,
                       "Total Criticals": total_criticals,
                       "Total Recoveries": total_recoveries,
                       "Total Vaccinations": total_vaccinations,
                       "Total Vaccinated": total_vaccinated,
                       "Total Vaccines Distributed": total_vaccines_distributed})
    
```

```python
covid_data_df = pd.DataFrame(covid_data)
```

```python
covid_data_df
```

```python
cleaned_covid_data_df = covid_data_df.dropna(subset=['Date'])
```
